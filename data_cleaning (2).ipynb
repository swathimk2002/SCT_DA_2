```python
# Extract zipfile

import zipfile
with zipfile.ZipFile("Globalsuperstore_dataset.zip") as zip_ref:
  zip_ref.extractall("data_folder")

# See the extracted files

import os
os.listdir("data_folder")

# Install the required package
!pip install xlrd


# Load the dataset
import pandas as pd
df = pd.read_excel("data_folder/Global Superstore.xls")


# To show the first few rows of the data so we can quickly see and check it.
df.head()


# To show a summary of the dataset like columns, data types, and missing values.
df.info()


# To identify the missing values
df.isnull().sum()

# to handle missing values
# Drop Postal Code
df = df.drop(columns=['Postal Code'])



# To drop duplicates
df =df.drop_duplicates()

# Convert datatypes,string column to a date format
df.columns = df.columns.str.strip().str.lower()

df['order date'] = pd.to_datetime(df['order date'])

df['ship date'] = pd.to_datetime(df['ship date'])

df.columns = df.columns.str.title()

# check

df.info()
df.head()



# save as csv file
df.to_csv("Cleaned_Global_Superstore.csv",index=False)



# to download the csv file

from google.colab import files
files.download("Cleaned_Global_Superstore.csv")
```
